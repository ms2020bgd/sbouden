```python
import pandas as pd
import numpy as np
pop = pd.read_csv('population_communes.csv')
surf = pd.read_csv('surface_departements.csv')
```

```python
pop = pop.rename(columns={
    'Code région': 'code_region',
    'Code département': 'code_dept',
    'Code canton': 'code_cant',
    'Code arrondissement':'code_arr',
    'Code commune':'code_comm',
    'Nom de la commune':'nom_comm',
    'Nom de la région':'nom_reg'
})
pop.head()
```

##### La somme des "Population municipale" du département

```python
df=pop[pop['variable'] == 'Population municipale'].copy()
df=df.groupby(['code_dept','code_region']).sum().reset_index()[['code_dept', 'code_region','value']]
df=df.rename(columns={'value': 'Population_municipale'})
df.head()
```

##### La somme des "Population totale" du département

```python
df_pop_tot=pop[pop['variable'] == 'Population totale'].copy()
df_pop_tot=df_pop_tot.groupby(['code_dept','code_region']).sum().reset_index()[['code_dept', 'code_region','value']]
df_pop_tot=df_pop_tot.rename(columns={'value': 'Population_totale'})
df_pop_tot.head()
```

##### La somme des "Population totale/municipale" du département

```python
tot=df_pop_tot['Population_totale']
df1=pd.concat([df, tot], axis=1)
df1.head()
```

##### La part (en pourcentage) de la population municipale par rapport à la population totale et la part (en pourcentage) de la population (municipale) du département au sein de sa région

```python
df1['%pop_municipale'] = df1.Population_municipale / df1.Population_totale * 100
df1['%region'] = df1.Population_municipale / df1.groupby('code_region').transform('sum').Population_municipale * 100
df1.head()
```

```python
surf=surf.rename(columns={'code_insee': 'code_dept'})
surf.head()
```

#### La densité de la population (municipale) en nb d'habitants / km2

```python
df=pd.merge(df, surf, how='inner', on=None, left_on=None, right_on=None,
         left_index=False, right_index=False, sort=True,
         copy=True, indicator=False, validate=None)
df.head()
```
