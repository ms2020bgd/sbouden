```python
import pandas as pd
import numpy as np
#key=open('API_key.txt').read()
import requests
from IPy import IP
```

```python
products = pd.read_csv('products.csv', error_bad_lines=False, sep=';')
```

```python
products.head()
```

```python
products=products.rename(columns={'ip_address': 'ipadd'})
```

```python
pd = products.dropna(subset=['ipadd'])
products.shape
```

```python
def get_currency(ip):
    #url = "http://api.ipstack.com/"+ip+"?access_key="+key
    url= "http://www.geoplugin.net/json.gp?ip="+ip
    data = requests.get(url).json()
    return data["geoplugin_currencyCode"]
#ip="138.195.148.164"
#get_currency(ip)
```

```python
get_currency("58.90.204.239")
```

```python
from IPy import IP
def validite(ip):
    flag = False
    while True:
        try:
            x = IP(ip)
            #print(ip)
            flag = True
            break
        except ValueError:
            #print(np.NaN)
            flag = False
            break
    return flag
validite('127.0.0.100000')
```

```python
print(products['ipadd'][0])
```

```python
products["valid"] = [validite(ip) for ip in products["ipadd"]]
products.head()
```

```python
#products=products.dropna(subset=["validite"])

index_to_drop = products[products['valid'] == False ].index
products.drop(index_to_drop, inplace=True)
products
```

```python
del products['valid']
products.head(10)
```

```python
products["currency"]=[get_currency(ip) for ip in products["ipadd"]]
products = products[products.currency.notna()].reset_index()
products.head()
```

```python
#contain=products.infos.str.contains('contain')
#contains=products.infos.str.contains('contains')
#may=products.infos.str.contains('may')
#List=products.infos.str.split()
#may
```

```python
import re
products["infos"] = products["infos"].apply(lambda s: s.replace("and",''))
products["infos"] = products["infos"].apply(lambda s: s.replace("Ingredients:",''))
products["infos"] = products["infos"].apply(lambda s: s.replace("May",''))
products["infos"] = products["infos"].apply(lambda s: s.replace("contain",''))
products["infos"] = products["infos"].apply(lambda s: s.replace("Contains",''))
products["infos"] = products["infos"].apply(lambda s: s.replace(",",''))
#del products["infos2"]
#products["infos2"] = products["infos"].apply(lambda s: re.split('\s+|[,;.-]\s*', s.strip()))
```

```python
ing=products['infos'].str.split(expand=True).rename(columns=lambda x: f"ingredient_{x+1}")
ing.head()
```

```python
new = products["infos"].str.get_dummies(sep=" ")
new.head(5)
```

```python
products=products.join(new, how="left")
```

```python
splitted = products['price'].str.split(' ')
price = splitted.str[0]
currency = splitted.str[1]
products['price']=price

#products.loc[lambda products: products['currency2'].notna()]
#del products["currency2"]
products.head()
```

```python
def prix_euro(ip):
    url= "http://www.geoplugin.net/json.gp?ip="+ip
    data = requests.get(url).json()
    return data["geoplugin_currencyConverter"]
    #return float(data["geoplugin_currencyConverter"])
ip1="138.195.148.164"
ip2="41.227.50.120"
prix_euro(ip2)
```

```python
products['rate']=[prix_euro(ip) for ip in products["ipadd"]]
products.head()
```

```python
products['price_eur'] = products['rate'].astype('float') * products['price'].astype('float')
products
```
